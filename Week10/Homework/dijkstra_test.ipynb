```python
import numpy as np
import matplotlib.pyplot as plt
import json
import dijkstra as dj
```

```python
#calculate an average with degree 2 and 100 nodes with 100 iterations
m = 2
N = 100
itnum = 100
```

```python
#calculate the averages
averages = dj.calculate(m = m, N = N, itnum = itnum)
```

```python
#save the results
with open('shortrun1.txt', 'w') as convert_file:
     convert_file.write(json.dumps(averages))
```

```python
#plot the results
fig = plt.figure()
plt.plot(list(averages.keys()), list(averages.values()), 'o')
plt.ylabel("Average distance", fontsize = 16)
plt.xlabel("Source degree", fontsize = 16)
plt.show()
```
