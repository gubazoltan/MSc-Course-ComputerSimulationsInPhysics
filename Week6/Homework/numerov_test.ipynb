```python
import numpy as np
import matplotlib.pyplot as plt
import numerov as num
```

## Square potential

#### As a function of energy.

```python
#height of the potential
V0 = 2

#possible energy values
Es = np.linspace(0.01,3.,200)*V0

width = 2*0.5 #total width of the potential

#number of positions in the potential
x_num = 100
```

```python
#and carry out the numerov method for the energies
Rs, Ts = num.multi_scattering_energies(Es = Es, potential = num.square_potential, width = width, V0 = -1*V0, x_num = x_num)
```

```python
#get the analytical values for the same potential
ac = num.get_acurve_energies(Es = Es, V0 = -1*V0, width = width)
```

```python
num.plotter_square_energies(Es = Es, V0 = V0, Ts  = Ts, ac = ac, width = width, save = True)
```

#### As a function of potential depth.

```python
#possible depths of the potential well
V0s = np.linspace(0,4,200)

#total width of the potential
width = 2*0.5

#energy of the scattering state
E = 0.1

#number of positions in the potential
x_num = 100
```

```python
#and carry out the numerov method for the energies
Rs, Ts = num.multi_scattering_potentials(E = E, potential = num.square_potential, width = width, V0s = -1*V0s, x_num = x_num)
```

```python
#get the analytical values for the same potential
ac = num.get_acurve_potentials(E = E, V0s = -1*V0s, width = width)
```

```python
num.plotter_square_potentials(E = E, V0s = V0s, Ts  = Ts, ac = ac, width = width, save = True)
```
