```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import spinchain_dynamics_v2 as spd
```

```python
N = 3
J = 0.25
h = 1
gammat = 2
esplit = 0.2
w = 1
Beta = 10
dtnum = 1000
tintnum = 10

params = N, J, h, gammat, esplit, w, Beta, dtnum, tintnum
```

```python
magns = spd.lindsolve(params)
```

```python
fig = plt.figure()
plt.plot(magns[:,0,1])
plt.plot(magns[:,1,1])
plt.plot(magns[:,2,1])

plt.xlabel("time [s]", fontsize = 14)
plt.ylabel(r"$M$", fontsize = 14)
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot3D(magns[:,0,0],magns[:,0,1],magns[:,0,2])
```
