```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import spinchain_dynamics as spd
```

### Let's test the code for some simple cases

First test the code on a chain of 

- $N=3$ spins
- no spin-spin coupling, $J=0$
- no external field, $h = 0$
- large gap so that there is no uphill relaxation

Expected outcome: the third spin relaxes in the y direction

```python
N = 3
J = 0
h = 0
gammat = 1
esplit = 10
w = 1
Beta = 1
dtnum = 100
tintnum = 4

params = N, J, h, gammat, esplit, w, Beta, dtnum, tintnum
```

```python
magns = spd.lindsolve(params)
```

```python
fig, axs = plt.subplots(1,3,figsize = (21,5))

for i in range(3):
    ax = axs[i]
    direction = i
    directs = ["x","y","z"]

    ax.plot(np.arange(dtnum*tintnum)/dtnum,magns[:,0,direction], label = "1")
    ax.plot(np.arange(dtnum*tintnum)/dtnum,magns[:,1,direction], label = "2")
    ax.plot(np.arange(dtnum*tintnum)/dtnum,magns[:,2,direction], label = "3")

    ax.set_xlabel(r"time [$\frac{2 \pi }{ \omega}$]", fontsize = 18)
    ax.set_ylabel(f"$M_{directs[direction]}$", fontsize = 18)
    ax.legend(fontsize = 14)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot3D(magns[:,2,0],magns[:,2,1],magns[:,2,2]);
```
